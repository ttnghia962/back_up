## Matplotlib
> Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. It was introduced by John Hunter in the year 2002. One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram, etc. 
- [plot()](#plot)
- [bar()](#bar)
- [scatter()](#scat)
- [figure()](#fig)

```python
import matplotlib.pyplot as plt
```

```python
""" Data given"""
x = [2, 4, 7, 9, 11]
y = [5, 1, 3, 4, 2]
```

<a id= 'plot'></a>
***

```python
# plot: create line plots

plt.plot(x, y)
plt.show()
```

```python
#Customization:

plt.plot(x, y, color='blue', linestyle='dashed', marker='o', label='Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')
plt.legend()
plt.show()
```

<a id= 'bar'></a>
***


```python
# bar: create bar plots

plt.bar(x, y)
plt.show()
```

```python
#Customization:

plt.bar(x,y,  color=['red', 'green', 'blue', 'purple', 'yellow'], width=0.5, tick_label=['A', 'B', 'C', 'D', 'E'])
# Customization options
plt.title('Customized Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.ylim(0, 10)           # Set y-axis limits
plt.show()
```

<a id = 'scat' > </a>
***

```python
# scatter: create scatter plots

plt.scatter(x, y)
plt.show()
```

```python
#Customization:

plt.scatter(x, y, color=['red', 'green', 'blue', 'purple', 'yellow'], s=[50, 100, 150, 200, 250], marker='x', edgecolors= 'red')
# Customization options
plt.title('Customized Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlim(0,12)
plt.ylim(0,6)

plt.show()
```

<a id = 'fig' > </a>

```python
# create a figure
fig = plt.figure(figsize=(8,5))
ax = plt.axes(projection='3d')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# To d liu mu
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# V biu 
plt.plot(x, y)

# Ty chnh nh trc x
new_ticks = [0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi]
new_labels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']

plt.xticks(new_ticks, new_labels, rotation=45, fontsize=10, color='red')

# Hin th biu 
plt.show()
```
